## Mendefinisikan fungsi - fungsi yang akan dipakai
Rumus untuk mendapatkan hambatan (R) adalah Voltase (V) dibagi dengan arus (I). Sedangkan rumus untuk daya (P) adalah Voltase (V) dikalikan ke arus (I)

```python
def hambatan(V, I):
    R = V / I
    return R

def daya(V, I):
    P = V*I
    return P
```

## Menambahkan kolom baru

Menambahkan kolom baru yang berisi hasil dari perhitungan fungsi yang barusan dibuat. Kolom baru adalah 'hambatan' dan 'daya'.

```python
df['hambatan'] = hambatan(df['V_m'],df['Current (A)'])
df['daya'] = daya(df['V_m'],df['Current (A)'])
```

```python
df
```

## Loop

Pertama-tama saya menginisialisasikan dataframe kosng untuk resistensi dan daya. Kemudian, bila nilai di kolom 'Cycle' adalah ganjil maka append setiap baris ke dataframe resistensi. Bila nilai di kolom 'Cycle' adalah genap maka append setiap baris ke dataframe daya

```python
resistensi = pd.DataFrame()
daya = pd.DataFrame()

for i, row in df.iterrows():
    if row['Cycle']%2 !=0:
        resistensi = resistensi.append(row,ignore_index=True)
    if row['Cycle']%2 ==0:
        daya = daya.append(row,ignore_index=True)
```
