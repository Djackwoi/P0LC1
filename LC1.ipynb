```python
# select yang hanya memiliki cycle genap saja
d1 = df[df['cycle'].isin([2,4,6,8])]
d1
```

Pada data diatas saya memilih kolom cycle yang hanya angka genap saja dengan method isin.

```python
# menghitung daya batrai cycle genap
v = d1['volt']
i = d1['current']

def hambatan():
    return (v*i)
 
print(hambatan())
```

Untuk menghitung daya batrai pada cycle genap maka saya mendevinisikan v dari dataframe d1 sbagai volt dan i sebagai current lalu setelah itu saya membuat fungsi dimana volt dikali dengan kuat arus atau current.

```python
# select yang hanya memiliki cycle ganjil saja saja
d2 = df[df['cycle'].isin([1,3,5,7,9])]
d2
```

sama seperti sebelumnya pada data diatas saya memilih kolom cycle yang hanya angka ganjil saja dengan method isin.

```python
# menghitung hambatan cycle ganjil
v = d2['volt']
i = d2['current']

def hambatan():
    return (v / i)
 
print(hambatan())
```

Untuk menghitung hambatan pada cycle ganjil maka saya mendevinisikan v dari dataframe d2 sbagai volt dan i sebagai current lalu setelah itu saya membuat fungsi dimana volt dibagi dengan kuat arus atau current.
