# 5. Answer

```python
def hambatan(V, I):
  "Function untuk formula Hambatan"
  R = V / I # R Kapasitas(ohm), V Tegangan(volt), I Kuat arus (Ampere)
  return R
```

```python
def daya(V, I):
  "Function untuk formula Daya"
  P = V * I # P Daya(watt), V Tegangan(volt), I Kuat arus (Ampere)
  return P
```

# 6. Jawaban

```python
for ind in df.index: # looping semua data yang ada di Dataset NMC1-9.xlsx
  v = df['V_m'][ind] # assign value di table 'V_m' ke variable v
  i = df['Current (A)'][ind] # assign value di table 'Current (A)' ke variable i
  cycle = df['Cycle'][ind] # assign value di table 'Cycle' ke variable cycle
  if cycle % 2 == 0: # Jika cycle adalah genap:
    p = daya(v, i,) # memangil function daya lalu assign ke variable p
    print('Tegangan: ', v, ', Kuat Arus: ', i,', siklus: ', cycle, ', Daya: ',p)
  else: # Jika ganjil:
    r = hambatan(v, i) # memangil function hambatan lalu assign ke variable r
    print('Tegangan: ', v, ', Kuat Arus: ', i,', siklus: ', cycle, ' Hambatan: ',r)
```
