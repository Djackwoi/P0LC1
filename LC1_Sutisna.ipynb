1. buat fungsi untuk menghitung hambatan

```python
def hambatan(v, i, t):
    return v / i / t
```

```python
def hambatan2():
    df["K_m"] = df.apply(lambda row : hambatan(row["V_m"], row["Current (A)"], row["time(s)"]), axis = 1)
    print(df)
    
if __name__ == '__hambatan2__':
    hambatan2()
    
hambatan2()
```

buat fungsi untuk menghitung daya

```python
def daya(v, i, t):
    return v * i / t
```

```python
def daya2():
    df["P_m"] = df.apply(lambda row : daya(row["V_m"], row["Current (A)"], row["time(s)"]), axis = 1)
    print(df)

if __name__ == '__daya2__':
    daya2()
daya2()
```
