```python
ganjil = df[df['Cycle'] % 2 == 1]
#Query untuk cycle ganjil
ganjil
```

Hasil

```python
# define function
def resistance (v,i):
    r = v / i
    return r

# List kosong
resistance_list = []

# For loop
for i in range(len(ganjil['time(s)'])):
    r = resistance(ganjil['V_m'].iloc[i], ganjil['Current (A)'].iloc[i])
    resistance_list.append(r)

ganjil['resistance'] = resistance_list

ganjil
```

# ## Menghitung setiap power baterai tiap waktunya untuk cycle genap

```python
genap = df[df['Cycle'] % 2 == 0]
#Query hanya untuk cycle genap
genap
```

```python
# define function
def power (v,i):
    r = v*i
    return r

# List kosong
power_list = []

# For loop
for i in range(len(genap['time(s)'])):
    r = power(genap['V_m'].iloc[i], genap['Current (A)'].iloc[i])
    power_list.append(r)

genap['power'] = power_list

genap
```
